## Training data

```python
td = TrainingGaussData()

# Import
params0_traj = Params0GaussTraj.fromFile("cache/cache_filtered.txt", nv=2)
dparams0_traj = DParams0GaussTraj.fromFile("cache/cache_derivs.txt", nv=2)

pt = int(0.5*len(params0_traj.params0_traj))

# Training data
td.reap_params0_traj_for_inputs(
    params0_traj=params0_traj, 
    tpt_start_inc=0,
    tpt_end_exc=pt,
    data_type=DataTypeGauss.TRAINING, 
    non_zero_idx_pairs_vv=non_zero_idx_pairs_vv
    )
td.reap_dparams0_traj_for_outputs(
    dparams0_traj=dparams0_traj, 
    tpt_start_inc=0,
    tpt_end_exc=pt,
    data_type=DataTypeGauss.TRAINING
    )
td.reap_params0_traj_for_inputs(
    params0_traj=params0_traj, 
    tpt_start_inc=pt,
    tpt_end_exc=params0_traj.nt-1, # Take off one to match derivs
    data_type=DataTypeGauss.VALIDATION, 
    non_zero_idx_pairs_vv=non_zero_idx_pairs_vv
    )
td.reap_dparams0_traj_for_outputs(
    dparams0_traj=dparams0_traj, 
    tpt_start_inc=pt,
    tpt_end_exc=dparams0_traj.nt,
    data_type=DataTypeGauss.VALIDATION
    )
```

```python
# Normalize outputs
td.calculate_output_standardizations_and_apply(0.2)

td.write_output_standardizations("cache")
```

```python
# Normalize inputs

# Reaction layer to use for the normalization
rxn_lyr_norm = RxnInputsGaussLayer.construct_one_init(
    nv=2,
    nh=1,
    freqs=np.array([freqs[-1]]),
    non_zero_idx_pairs_vv=non_zero_idx_pairs_vv,
    non_zero_idx_pairs_hv=non_zero_idx_pairs_hv,
    non_zero_idx_pairs_hh=non_zero_idx_pairs_hh,
    rxn_specs=rxn_specs
    )

# Normalize inputs
model.calculate_rxn_normalization(
    rxn_lyr=rxn_lyr_norm, 
    inputs=td.train_inputs, 
    percent=0.2
    )
```

```python
print("Reaction normalization: Mean: ", model.rxn_mean)
print("Reaction normalization: Std dev: ", model.rxn_std_dev)
```

# Train

```python
loss_fn = tf.keras.losses.MeanSquaredError()
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt,
              loss=loss_fn,
              run_eagerly=False)

# Log outputs for tensorboard
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, 
    histogram_freq=1,
    write_graph=False
    )

# Write weights of lowest validation score
val_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'trained/weights_lowest_val', 
    monitor='val_loss', 
    verbose=1,
    save_best_only=True, 
    save_weights_only=True,
    mode='auto', 
    save_frequency=1
    )

# Train!
model.fit(
    td.train_inputs, 
    td.train_outputs_stdrd, 
    validation_data=(td.valid_inputs,td.valid_outputs_stdrd),
    epochs=25, 
    callbacks=[tensorboard_callback,val_checkpoint],
    batch_size=8
)

# Save
model.save("trained/trained_final", save_traces=False)
```

## Accuracy

```python
def make_input_single(idx, train_inputs):
    in0 = {}
    for key,val in train_inputs.items():
        in0[key] = np.array([val[idx]])
    return in0

def make_output_single(idx, train_outputs_norm):
    out0_norm_true = {}
    for key,val in train_outputs_norm.items():
        out0_norm_true[key] = np.array([val[idx]])
    return out0_norm_true

def make_input_multiple(idxs, train_inputs):
    in0 = {}
    for key,val in train_inputs.items():
        in0[key] = np.array(val[idxs])
    return in0

def make_output_multiple(idxs, train_outputs_norm):
    out0_norm_true = {}
    for key,val in train_outputs_norm.items():
        out0_norm_true[key] = np.array(val[idxs])
    return out0_norm_true

def test_output_np(in0):
    out0_norm = model(in0)
    for key,val in out0_norm.items():
        out0_norm[key] = val.numpy()
    return out0_norm
```
